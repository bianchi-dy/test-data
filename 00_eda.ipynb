```python
clean["job_tags"] = clean.apply(lambda r: r["Job Tags"].split(","), axis=1)
clean
```

```python
unique_tags = clean["job_tags"].to_list()
unique_tags
```

```python
flattened = [val for sublist in unique_tags for val in sublist]
unique_tags = list(set(flattened))
unique_tags.sort()
unique_tags
```

```python
at = clean["Archetype"].unique()
at
```

### Get archetype counts

```python
d = {}

for u in unique_tags:
    d[u] = [0 for i in range(len(at))]
    
d["archetype"] = at
d
```

```python
industry_archetypes = pd.DataFrame(d)
industry_archetypes
```

```python
industry_archetypes["total"] = 0
```

```python
for r, row in tqdm(industry_archetypes.iterrows()):
    search = row["archetype"]
    subdf = clean[clean["Archetype"] == search]
    for u in unique_tags:
        for k, krow in subdf.iterrows():
            if u in krow["job_tags"]:
                print (r, u)
                industry_archetypes.at[r, u] +=1
            
    industry_archetypes.at[r, "total"] = len(subdf)
    display(industry_archetypes)
```

```python
from collections import Counter
counter_object = Counter(flat_signals)
keys = counter_object.keys()
num_values = counter_object.values()
df_signals = pd.DataFrame({"signal": keys, "value": num_values}).sort_values("value")
df_signals
```

## STEEP ranking

```python
steep_ranking = clean.groupby(["STEEP Change 1"])["id"].count().reset_index()
sr1 = steep_ranking.rename(columns={"STEEP Change 1": "change"}).sort_values("id")
sr1["cat"] = "STEEP Change 1"
sr1
```

```python
steep_ranking = clean.groupby(["STEEP Change 2"])["id"].count().reset_index()
sr2 = steep_ranking.rename(columns={"STEEP Change 2": "change"}).sort_values("id")
sr2["cat"] = "STEEP Change 2"
sr2
```

```python
steep_ranking = clean.groupby(["STEEP Change 3"])["id"].count().reset_index()
sr3 = steep_ranking.rename(columns={"STEEP Change 3": "change"}).sort_values("id")
sr3["cat"] = "STEEP Change 3"
sr3
```

```python
srs = pd.concat([sr1, sr2, sr3]).sort_values("id")
srs.to_csv("data/steep-change-counts.csv")
```
